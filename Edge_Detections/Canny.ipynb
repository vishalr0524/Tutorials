```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/home/hp/Documents/Daily_Task/Day_2/Assets/sudoku.jpg')      # use your file name
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
plt.show()
```

```python
blur = cv2.GaussianBlur(gray, (5,5), 1.4)   # kernel size 55, sigma=1.4

plt.imshow(blur, cmap='gray')
plt.title("After Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
# Apply Canny edge detection
edges = cv2.Canny(blur, 50, 150)  # low=50, high=150 are typical starter values

plt.imshow(edges, cmap='gray')
plt.title("Canny Edges (50,150)")
plt.axis('off')
plt.show()
```

```python
edges_low = cv2.Canny(blur, 30, 100)   # more sensitive
edges_high = cv2.Canny(blur, 100, 200) # less sensitive

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(edges_low, cmap='gray')
plt.title("Canny (30,100)  More edges")

plt.subplot(1,2,2)
plt.imshow(edges_high, cmap='gray')
plt.title("Canny (100,200)  Fewer edges")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read and preprocess
img = cv2.imread("/home/hp/Documents/Daily_Task/Day_2/Assets/sudoku.jpg", cv2.IMREAD_GRAYSCALE)
blur = cv2.GaussianBlur(img, (5,5), 1.4)

# Sobel
sobelx = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=3)
sobel_mag = cv2.magnitude(sobelx, sobely)
sobel_edges = np.uint8(np.clip(sobel_mag, 0, 255))

# Canny
canny_edges = cv2.Canny(blur, 50, 150)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(sobel_edges, cmap='gray')
plt.title("Sobel Edge Detection")

plt.subplot(1,2,2)
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.show()
```
